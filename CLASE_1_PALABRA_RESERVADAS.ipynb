Palabras Reservadas

Las palabras reservadas en Python (o keywords) son palabras que el lenguaje ya tiene definidas y que no puedes usar como nombres de variables, funciones, clases, etc. porque cumplen una función especial dentro del código.

En otras palabras son palabras con un significado propio dentro de Python y no se pueden modificar.

Características de las palabras reservadas
1.No pueden usarse como identificadores (variables, funciones, clases).

2.Son parte del núcleo del lenguaje, por lo que Python las interpreta de forma especial.

3.Todas están en minúsculas, excepto True, False y None.

4.Son limitadas y fijas, pero pueden aumentar con nuevas versiones de Python.

5.Definen la lógica del programa, como condiciones, ciclos, funciones, clases, excepciones, etc.

# Ejemplo práctico con varias palabras reservadas


```python
def suma(a, b):
    if a > 0 and b > 0:
        return a + b
    else:
        return 0

resultado = suma(4, 5)
print(resultado)
```
